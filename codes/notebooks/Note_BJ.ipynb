```python
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_folder = r'/home/youngsuk/repos/data'
df_ftse = pd.read_csv(os.path.join(data_folder, r'FTSE100.csv'), delimiter='\t', parse_dates=['Date']).set_index('Date')
df_gbpusd = pd.read_csv(os.path.join(data_folder,r'GBPUSD.csv'), delimiter='\t', parse_dates=['Date']).set_index('Date')
```

```python
df_merged = pd.merge(df_ftse, df_gbpusd, on='Date')
df_merged = df_merged[df_merged.index > '2019-07-15']
```

```python
fig, ax1 = plt.subplots()
ax1.plot(df_merged.index, df_merged['FTSE 100'], 'b')
ax2 = ax1.twinx()
ax2.plot(df_merged.index, df_merged['GBPUSD'], 'r')
fig.autofmt_xdate()
```
